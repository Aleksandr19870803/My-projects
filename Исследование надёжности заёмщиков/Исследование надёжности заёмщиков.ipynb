### Замена типа данных

```python
def number(count):
    if count>5:
        return count/10
    else:
        return count


def method(row):
    if row>18000:
        return row/100
    else:
        return row
  
data['days_employed'] = abs(data['days_employed'].apply(method).astype('int'))
data['total_income'] = data['total_income'].astype('int')
data['education'] = data['education'].str.lower()
data['children'] = abs(data['children'].apply(number).astype('int'))
```

Но ты можешь обратить внимание, что стаж все равно получается нереально большой. Об этом необходимо сообщить разработчикам базы данных во избежании дальнейших неприятностей. ### Лемматизация

```python

data['purpose'].unique()
```

```python
from pymystem3 import Mystem
m = Mystem() 
data['lemmas'] = data['purpose'].apply(m.lemmatize)
```

```python
def word(i):
    if 'жилье' in i:
        return 'жилье'
    if 'автомобиль' in i:
        return 'автомобиль'
    if 'образование' in i:
        return 'образование'
    if 'свадьба' in i:
        return 'свадьба'
    if 'недвижимость' in i:
        return 'недвижимость'
    
        
data['lemmas_1'] = data['lemmas'].apply(word) 
```

**Вывод**

Выведем уникалные значения столбца цель получения кредита.Проведем лемматизацию столбца цель получения кредита.Выведем новый столбец с леммами из столбца цель получения кредита.Из столбца с леммами выделим функцией word ключевые слова для столбца lemmas_1, по нему и будем делать сводную таблицу.

### Категоризация данных

```python
def income(level):
    if level <160000:
        return 'низкий'
    if 300000 > level> 160000:
        return 'средний'
    else:
        return 'высокий'
data['level_income'] = data['total_income'].apply(income) 
```
